```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("/content/01.Data Cleaning and Preprocessing.csv")

# Convert 'Observation' column to a datetime or remove it, depending on your analysis needs
# If you need it as datetime:
# df['Observation'] = pd.to_datetime(df['Observation'], format='%d-%H:%M')

# If you want to exclude it from numerical analysis:
df = df.drop('Observation', axis=1)

# 1. Distribution of variables
plt.figure(figsize=(15, 10))
df.hist(bins=20, figsize=(20, 15), color='skyblue')
plt.suptitle('Distribution of Variables', fontsize=16)
plt.show()

# 2. Boxplot to identify outliers
plt.figure(figsize=(20, 10))
sns.boxplot(data=df, palette='Set3')
plt.title('Boxplots of Variables to Identify Outliers', fontsize=16)
plt.xticks(rotation=90)
plt.show()

# 3. Correlation heatmap to check for correlations between variables
plt.figure(figsize=(15, 10))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix', fontsize=16)
plt.show()
```
