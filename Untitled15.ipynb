```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
file_path = '/content/01.Data Cleaning and Preprocessing.csv'  # Replace with the correct path to your CSV file
df = pd.read_csv(file_path)

# Basic data overview
print(df.head())
print(df.describe())
print(df.dtypes)

# Select only numeric columns for correlation
numeric_df = df.select_dtypes(include=['float64', 'int64'])

# 3. Correlation heatmap to check for correlations between variables
plt.figure(figsize=(15, 10))
correlation_matrix = numeric_df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix', fontsize=16)
plt.show()

# 1. Histograms for each numerical column
df.hist(bins=15, figsize=(15, 10), grid=False)
plt.suptitle('Distribution of Variables', fontsize=16)
plt.show()

# 2. Boxplots to identify outliers
plt.figure(figsize=(15, 10))
sns.boxplot(data=df, orient="h")
plt.title('Boxplot of Variables to Identify Outliers', fontsize=16)
plt.show()

# 3. Correlation heatmap to check for correlations between variables
plt.figure(figsize=(15, 10))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix', fontsize=16)
plt.show()
```
