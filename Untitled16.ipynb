```python

# Load the dataset
df = pd.read_csv('heart.csv')

# Checking for missing values
missing_values = df.isnull().sum()
missing_values
```

```python
df.describe()
```

```python
import numpy as np

# Z-score for outlier detection
from scipy import stats

# Select the columns for numeric data
numeric_columns = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']

# Calculate Z-scores
z_scores = np.abs(stats.zscore(df[numeric_columns]))

# Define a threshold (e.g., Z-score > 3 indicates outliers)
threshold = 3

# Get the outliers
outliers = np.where(z_scores > threshold)

# Remove outliers
df_cleaned = df[(z_scores < threshold).all(axis=1)]

print(f"Number of rows before removing outliers: {df.shape[0]}")
print(f"Number of rows after removing outliers: {df_cleaned.shape[0]}")
```

1:What is the distribution of age?

```python
import matplotlib.pyplot as plt
df['age'].plot(kind='hist', bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

2:Does gender (sex) affect the likelihood of having heart disease?

```python
import seaborn as sns
sns.countplot(x='target', hue='sex', data=df)
plt.title('Heart Disease Prevalence by Gender')
plt.xlabel('Heart Disease (1 = Disease, 0 = No Disease)')
plt.ylabel('Count')
plt.legend(['Female', 'Male'], loc='upper right')
plt.show()
```

3: What is the relationship between cholesterol levels and heart disease?

```python
sns.boxplot(x='target', y='chol', data=df)
plt.title('Cholesterol Levels vs. Heart Disease')
plt.xlabel('Heart Disease (1 = Disease, 0 = No Disease)')
plt.ylabel('Cholesterol Level')
plt.show()
```

 4: How does chest pain type (cp) relate to heart disease?

```python
sns.countplot(x='cp', hue='target', data=df)
plt.title('Chest Pain Type and Heart Disease')
plt.xlabel('Chest Pain Type')
plt.ylabel('Count')
plt.legend(['No Disease', 'Heart Disease'])
plt.show()
```

5: How does maximum heart rate achieved (thalach) vary between patients with and without heart disease?


```python
sns.boxplot(x='target', y='thalach', data=df)
plt.title('Max Heart Rate Achieved vs. Heart Disease')
plt.xlabel('Heart Disease (1 = Disease, 0 = No Disease)')
plt.ylabel('Maximum Heart Rate')
plt.show()
```

6: What is the relationship between fasting blood sugar (fbs > 120 mg/dl) and heart disease?

```python
sns.countplot(x='fbs', hue='target', data=df)
plt.title('Fasting Blood Sugar vs. Heart Disease')
plt.xlabel('Fasting Blood Sugar > 120 mg/dl (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.legend(['No Disease', 'Heart Disease'])
plt.show()
```

7: What is the correlation between various features and heart disease?


```python
corr = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Features')
plt.show()
```
