```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset from the CSV file
df = pd.read_csv('/content/01.Data Cleaning and Preprocessing.csv')

# Display the first few rows to verify the data
print(df.head())

```

```python
# Plotting a bar chart and a line chart
plt.figure(figsize=(14, 7))

# Bar chart for 'Y-Kappa'
plt.bar(df['Observation'], df['Y-Kappa'], color='blue', alpha=0.6, label='Y-Kappa')
```

```python
# Line chart for 'ChipRate'
plt.plot(df['Observation'], df['ChipRate'], color='red', marker='o', label='ChipRate')

# Adding labels, title, and legend
plt.xlabel('Observation')
plt.ylabel('Values')
plt.title('Y-Kappa and ChipRate Over Observations')
plt.legend()

# Rotating x-ticks for better readability
plt.xticks(rotation=45)

# Display the plot
plt.tight_layout()
plt.show()
```
